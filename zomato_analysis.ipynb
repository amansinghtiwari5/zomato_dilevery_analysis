```python
df['Delivery_person_Age'] = df['Delivery_person_Age'].fillna(df['Delivery_person_Age'].median())
```

```python
df['Delivery_person_Ratings'] = df['Delivery_person_Ratings'].fillna(df['Delivery_person_Ratings'].median())
```

```python
df['multiple_deliveries'] = df['multiple_deliveries'].fillna(0)
```

```python
df['Weather_conditions'] = df['Weather_conditions'].fillna(df['Weather_conditions'].mode()[0])
```

```python
df['Road_traffic_density'] = df['Road_traffic_density'].fillna(df['Road_traffic_density'].mode()[0])
```

```python
df['Festival'] = df['Festival'].fillna('No')
```

```python
df['City'] = df['City'].fillna('Unknown')
```

```python
df = df.dropna(subset=['Time_Orderd'])
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

⏱️ Distribution of Delivery Time

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Time_taken (min)'], bins=30, kde=True)
plt.title("Distribution of Delivery Time (Minutes)")
plt.xlabel("Time Taken (min)")
plt.ylabel("Frequency")
plt.show()
```

Impact of Traffic on Delivery Time

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Road_traffic_density', y='Time_taken (min)', data=df)
plt.title("Delivery Time vs Road Traffic Density")
plt.show()
```

Weather Conditions vs Delivery Time

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Weather_conditions', y='Time_taken (min)', data=df)
plt.xticks(rotation=30)
plt.title("Weather Impact on Delivery Time")
plt.show()
```

Festival vs Non-Festival Days 

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Festival', y='Time_taken (min)', data=df)
plt.title("Festival Impact on Delivery Time")
plt.show()
```

City-wise Delivery Performance

```python
city_avg = df.groupby('City')['Time_taken (min)'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
city_avg.plot(kind='bar')
plt.title("Average Delivery Time by City")
plt.ylabel("Avg Time (min)")
plt.show()
```

Delivery Person Rating vs Delivery Time

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='Delivery_person_Ratings',
    y='Time_taken (min)',
    data=df,
    alpha=0.4
)
plt.title("Delivery Time vs Delivery Person Ratings")
plt.show()
```

Multiple Deliveries Impact

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='multiple_deliveries', y='Time_taken (min)', data=df)
plt.title("Impact of Multiple Deliveries on Time")
plt.show()
```
